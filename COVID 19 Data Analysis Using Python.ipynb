#  Calculate maximum infection rate for all of the countries.

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
confirmed_cases = corona_dataset_aggregated.columns['10/22/20']
for country in countries :
    max_infection_rates.append(corona_dataset_aggregated.loc[country].diff().max())
corona_dataset_aggregated['Maximum Infection Rate'] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

# Create a new data frame

```python
corona_data = pd.DataFrame(corona_dataset_aggregated['Maximum Infection Rate'])
corona_data.head(10)
```

# Importing World Happiness Report dataset 

```python
world_happiness_report = pd.read_csv("world_happiness_report_2020.csv")
world_happiness_report.head()
```

```python
world_happiness_report.shape
```

```python
columns_to_dropped = ['Generosity','Perceptions of corruption', 'Regional indicator', 'upperwhisker', 'lowerwhisker', 'Ladder score', 'Standard error of ladder score']
world_happiness_report.drop(columns_to_dropped,axis=1 , inplace=True)
```

```python
world_happiness_report.head()
```

```python
columns_to_dropped = ['Ladder score in Dystopia','Explained by: Log GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy', 'Explained by: Freedom to make life choices', 'Explained by: Generosity', 'Explained by: Perceptions of corruption', 'Dystopia + residual']
world_happiness_report.drop(columns_to_dropped,axis=1 , inplace=True)
```

```python
world_happiness_report.set_index(['Country name'],inplace=True)
world_happiness_report.head()
```
